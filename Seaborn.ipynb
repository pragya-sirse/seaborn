```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly as px
```

## Seaborn

```python
# relational plot
sns.load_dataset('titanic')
tips=sns.load_dataset('tips')
tips
```

```python
# axis level function
sns.scatterplot(data=tips, x='tip',y='total_bill')
```

```python
# relplot is a figure level function
sns.relplot(data=tips, x='tip' , y='total_bill',kind='scatter',hue='sex',style='time',size='size')
```

```python
sns.relplot(data=tips, x='tip' , y='total_bill',kind='line',hue='sex',style='time',size='size')
```

```python
gap=px.data.gapminder()
temp_df=gap[gap['country']=='India']
temp_df
```

```python
sns.lineplot(data=temp_df, x='year',y='lifeExp')
```

```python
sns.relplot(data=temp_df,x='year',y='lifeExp' ,kind='line')
```

```python
# hue
temp_df=gap[gap['country'].isin(['India','Pakistan','China'])]
temp_df
```

```python
sns.relplot(data=temp_df,x='year',y='lifeExp' ,hue='country',kind='line')
```

```python
sns.lineplot(data=temp_df,x='year',y='lifeExp' ,hue='country')
```

```python
temp_df=gap[gap['country'].isin(['India','Brazil','Germany'])]
temp_df
```

```python
sns.relplot(data=temp_df,x='year',y='lifeExp',hue='country',kind='line',style='continent')
```

```python
sns.lineplot(data=temp_df,x='year',y='lifeExp',hue='country')
```

```python
# facet plot
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',col='sex',row='day')
```

```python
# col wrap
sns.relplot(data=gap,x='lifeExp',y='gdpPercap',kind='scatter',col='year',col_wrap=4)
```

```python
# Distribution plot -> histogram
sns.histplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill',kind='hist',bins=50)
```

```python
sns.displot(data=tips,x='day',kind='hist',bins=50)
```

```python
sns.displot(data=tips,x='tip',kind='hist',bins=50,hue='sex')
```

```python
sns.displot(data=titanic,x='age',kind='hist',bins=50,col='sex')
```

```python
# Matrix Plot
plt.figure(figsize=(20,10))
temp_df=gap.pivot(index='country',columns='year',values='lifeExp')
temp_df
sns.heatmap(temp_df)
```

```python
temp_df=gap[gap['continent']=='Asia'].pivot(index='country',columns='year',values='lifeExp')
plt.figure(figsize=(20,10))
sns.heatmap(temp_df,annot=True,linewidths=0.5,cmap='summer')
```
